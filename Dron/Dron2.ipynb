```python
import cv2
from tracker import *
import numpy as np

# Create tracker object
tracker = EuclideanDistTracker()

cap = cv2.VideoCapture("Trim.mp4")

# Object detection from Stable camera
object_detector = cv2.createBackgroundSubtractorMOG2()
```

```python
while True:
    ret, frame = cap.read()
    #height, width, _ = frame.shape
    
    
    # Extract Region of interest
    roi = frame[200: 1080, 200: 800]

    # 1. Object Detection
    mask = object_detector.apply(roi)
    _, mask = cv2.threshold(mask, 254, 255, cv2.THRESH_BINARY)
    

    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    detections = []
    for cnt in contours:
        # Calculate area and remove small elements
        area = cv2.contourArea(cnt)

        if (area > 3500) and (area < 4500):
            #cv2.drawContours(roi, [cnt], -1, (0, 255, 0), 2)

            #cv2.drawContours(frame, cnt, -1, (200, 200, 0), 3)
            p = cv2.arcLength(cnt, True)
            num_versh = cv2.approxPolyDP(cnt, p*0.01, True)


            x, y, w, h = cv2.boundingRect(cnt)
            detections.append([x, y, w, h])

    # 2. Object Tracking
    boxes_ids = tracker.update(detections)
    for box_id in boxes_ids:
        x, y, w, h, id = box_id

        if (len(num_versh) >= 13) and (len(num_versh) <= 20):
            cv2.putText(frame, str(id), (x, y - 15), cv2.FONT_HERSHEY_PLAIN, 2, (255, 0, 0), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)

    cv2.imshow("roi", roi)
    cv2.imshow("Frame", frame)
    cv2.imshow("Mask", mask)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        cv2.destroyAllWindows()
        break
```
